# A top 5 bajnokság szezon végi adatai 2000 és 2023 között

Programcsomagok importálása

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
```

Webscraping az https://fbref.com oldaláról

```python

#url_df = 'https://fbref.com/en/comps/9/2022-2023/2022-2023-Premier-League-Stats'
start_year = 2000
end_year = 2023
fields = ['Rk', 'Squad', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts']
leagues = [('Premier-League', 9), ('La-Liga', 12), ('Serie-A', 11), ('Bundesliga', 20), ('Ligue-1', 13)]

lst = []

for league_name, league_id in leagues:
    for i in range(start_year, end_year):
        BASE ='https://fbref.com'
        URL = f'{BASE}/en/comps/{league_id}/{i}-{i+1}/{i}-{i+1}-{league_name}-Stats'
        print(URL)
        df = pd.read_html(URL)[0]
        df = df[fields]
        df = df.reset_index(drop=True)
        df.insert(0, "League", league_name)
        df.insert(0, "Season", i+1)
        lst.append(df)
        
        time.sleep(60/20)
    
```

Adatok összeolvasztása, majd exportálása csv fájlba

```python
big_df = pd.concat(lst, axis=0)
big_df.to_csv(f'A top 5 bajnokság {start_year}-{end_year} közötti adatai.csv', encoding = 'utf-8', index = False)

big_df.head(5)
```

Adatokat szétbontása bajnokságonként

```python
eng_df = big_df.loc[big_df['League'] == 'Premier-League']
spa_df = big_df.loc[big_df['League'] == 'La-Liga']
ita_df = big_df.loc[big_df['League'] == 'Serie-A']
ger_df = big_df.loc[big_df['League'] == 'Bundesliga']
fra_df = big_df.loc[big_df['League'] == 'Ligue-1']

eng_df.head(5)
```

Adatok szétbontása csapatonként és idényenként, példaszűrés az Arsenal csapat eredményeire

```python
eng_teams = eng_df.groupby('Squad')
eng_seasons = eng_df.groupby('Season')

eng_teams.get_group('Arsenal').head(5)
```
