# CBR kiszámítása a top 5 bajnokságban 2000 és 2023 között

Programcsomagok importálása

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Lokális adatbázis importálása

```python
df = pd.read_csv('A top 5 bajnokság 2000-2023 közötti adatai.csv')
df2 = df
```

Szükséges számított mezők létrehozása és kiszámítása

```python
#df2['WPCTi'] = (df2['W'] + (df2['D'] / 2)) / df2['MP']#
#df2['WPCTi'] = df2['W'] / df2['MP']
df2['WPCTi'] = df2['Pts'] / (df2['MP'] * 3)
df2
```

Adatok szétválasztása országonként

```python
eng_df = df2.loc[df2['League'] == 'Premier-League']
spa_df = df2.loc[df2['League'] == 'La-Liga']
ita_df = df2.loc[df2['League'] == 'Serie-A']
ger_df = df2.loc[df2['League'] == 'Bundesliga']
fra_df = df2.loc[df2['League'] == 'Ligue-1']
```

CBR értékek kiszámítása bajnokságonként

```python
eng_teams = eng_df.groupby(['Squad', 'Season'])['WPCTi'].sum().to_frame()
eng_teams['-WPCT'] = pow(eng_teams['WPCTi'] - eng_teams.groupby('Squad')['WPCTi'].transform(lambda x: x.mean()), 2)
eng_teams_SD = np.sqrt(eng_teams.groupby('Squad')['-WPCT'].sum() / eng_teams.groupby('Squad')['-WPCT'].count())
mean_eng_teams_SD = eng_teams_SD.sum() / eng_teams_SD.count()

eng_teams['-WPCT2'] = pow(eng_teams['WPCTi'] - 0.500, 2)
eng_seasons_SD = np.sqrt(eng_teams.groupby('Season')['-WPCT2'].sum() / eng_teams.groupby('Season')['-WPCT2'].count())
mean_eng_seasons_SD = eng_seasons_SD.sum() / eng_seasons_SD.count()

eng_CBR = mean_eng_teams_SD / mean_eng_seasons_SD



spa_teams = spa_df.groupby(['Squad', 'Season'])['WPCTi'].sum().to_frame()
spa_teams['-WPCT'] = pow(spa_teams['WPCTi'] - spa_teams.groupby('Squad')['WPCTi'].transform(lambda x: x.mean()), 2)
spa_teams_SD = np.sqrt(spa_teams.groupby('Squad')['-WPCT'].sum() / spa_teams.groupby('Squad')['-WPCT'].count())
mean_spa_teams_SD = spa_teams_SD.sum() / spa_teams_SD.count()

spa_teams['-WPCT2'] = pow(spa_teams['WPCTi'] - 0.500, 2)
spa_seasons_SD = np.sqrt(spa_teams.groupby('Season')['-WPCT2'].sum() / spa_teams.groupby('Season')['-WPCT2'].count())
mean_spa_seasons_SD = spa_seasons_SD.sum() / spa_seasons_SD.count()

spa_CBR = mean_spa_teams_SD / mean_spa_seasons_SD



ita_teams = ita_df.groupby(['Squad', 'Season'])['WPCTi'].sum().to_frame()
ita_teams['-WPCT'] = pow(ita_teams['WPCTi'] - ita_teams.groupby('Squad')['WPCTi'].transform(lambda x: x.mean()), 2)
ita_teams_SD = np.sqrt(ita_teams.groupby('Squad')['-WPCT'].sum() / ita_teams.groupby('Squad')['-WPCT'].count())
mean_ita_teams_SD = ita_teams_SD.sum() / ita_teams_SD.count()

ita_teams['-WPCT2'] = pow(ita_teams['WPCTi'] - 0.500, 2)
ita_seasons_SD = np.sqrt(ita_teams.groupby('Season')['-WPCT2'].sum() / ita_teams.groupby('Season')['-WPCT2'].count())
mean_ita_seasons_SD = ita_seasons_SD.sum() / ita_seasons_SD.count()

ita_CBR = mean_ita_teams_SD / mean_ita_seasons_SD



ger_teams = ger_df.groupby(['Squad', 'Season'])['WPCTi'].sum().to_frame()
ger_teams['-WPCT'] = pow(ger_teams['WPCTi'] - ger_teams.groupby('Squad')['WPCTi'].transform(lambda x: x.mean()), 2)
ger_teams_SD = np.sqrt(ger_teams.groupby('Squad')['-WPCT'].sum() / ger_teams.groupby('Squad')['-WPCT'].count())
mean_ger_teams_SD = ger_teams_SD.sum() / ger_teams_SD.count()

ger_teams['-WPCT2'] = pow(ger_teams['WPCTi'] - 0.500, 2)
ger_seasons_SD = np.sqrt(ger_teams.groupby('Season')['-WPCT2'].sum() / ger_teams.groupby('Season')['-WPCT2'].count())
mean_ger_seasons_SD = ger_seasons_SD.sum() / ger_seasons_SD.count()

ger_CBR = mean_ger_teams_SD / mean_ger_seasons_SD



fra_teams = fra_df.groupby(['Squad', 'Season'])['WPCTi'].sum().to_frame()
fra_teams['-WPCT'] = pow(fra_teams['WPCTi'] - fra_teams.groupby('Squad')['WPCTi'].transform(lambda x: x.mean()), 2)
fra_teams_SD = np.sqrt(fra_teams.groupby('Squad')['-WPCT'].sum() / fra_teams.groupby('Squad')['-WPCT'].count())
mean_fra_teams_SD = fra_teams_SD.sum() / fra_teams_SD.count()

fra_teams['-WPCT2'] = pow(fra_teams['WPCTi'] - 0.500, 2)
fra_seasons_SD = np.sqrt(fra_teams.groupby('Season')['-WPCT2'].sum() / fra_teams.groupby('Season')['-WPCT2'].count())
mean_fra_seasons_SD = fra_seasons_SD.sum() / fra_seasons_SD.count()

fra_CBR = mean_fra_teams_SD / mean_fra_seasons_SD


print(eng_CBR, ita_CBR, spa_CBR, ger_CBR, fra_CBR)
```
