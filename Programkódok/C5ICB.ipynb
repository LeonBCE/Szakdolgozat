# C5ICB kiszámítása a top 5 bajnokságra 2000 és 2023 között

Programcsomagok importálása

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Lokális adatbázis importálása

```python
df = pd.read_csv('A top 5 bajnokság 2000-2023 közötti adatai.csv')
df2 = df
```

Adatok szétválasztása országonként

```python
eng_df = df2.loc[df2['League'] == 'Premier-League']
spa_df = df2.loc[df2['League'] == 'La-Liga']
ita_df = df2.loc[df2['League'] == 'Serie-A']
ger_df = df2.loc[df2['League'] == 'Bundesliga']
fra_df = df2.loc[df2['League'] == 'Ligue-1']
```

Szükséges számított mezők létrehozása és kiszámítása

```python
eng_C5ICB = ((eng_df.loc[eng_df['Rk'] <= 5].groupby('Season')['Pts'].sum()
     / eng_df.groupby('Season')['Pts'].sum())
    / ( 5 / eng_df.groupby('Season')['Pts'].count())
    * 100).to_frame()

spa_C5ICB = ((spa_df.loc[spa_df['Rk'] <= 5].groupby('Season')['Pts'].sum()
     / spa_df.groupby('Season')['Pts'].sum())
    / ( 5 / spa_df.groupby('Season')['Pts'].count())
    * 100).to_frame()

ita_C5ICB = ((ita_df.loc[ita_df['Rk'] <= 5].groupby('Season')['Pts'].sum()
     / ita_df.groupby('Season')['Pts'].sum())
    / ( 5 / ita_df.groupby('Season')['Pts'].count())
    * 100).to_frame()

ger_C5ICB = ((ger_df.loc[ger_df['Rk'] <= 5].groupby('Season')['Pts'].sum()
     / ger_df.groupby('Season')['Pts'].sum())
    / ( 5 / ger_df.groupby('Season')['Pts'].count())
    * 100).to_frame()

fra_C5ICB = ((fra_df.loc[fra_df['Rk'] <= 5].groupby('Season')['Pts'].sum()
     / fra_df.groupby('Season')['Pts'].sum())
    / ( 5 / fra_df.groupby('Season')['Pts'].count())
    * 100).to_frame()
```

Adatok összekapcsolása, exportálása csv fájlba

```python
C5ICB = pd.concat([eng_C5ICB['Pts'], ita_C5ICB['Pts'], spa_C5ICB['Pts'], ger_C5ICB['Pts'], fra_C5ICB['Pts']],
                 axis = 1, keys = ['eng', 'ita', 'spa', 'ger', 'fra'])

C5ICB.to_csv(f'A top 5 bajnokság C5ICB értéke 2000 és 2023 között.csv', encoding = 'utf-8', index = False)

C5ICB.head(5)
```

C5ICB ábrák elkészítése és exportálása

```python
font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

#plt.xticks(rotation=90)

sns.lineplot(data = C5ICB, x = C5ICB.index, y = 'eng', color = 'black', marker = 'o')

plt.title('Premier League', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('C5ICB', **font)

plt.ylim(120, 170)
plt.tight_layout()

plt.savefig("eng_C5ICB.png")



fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

#plt.xticks(rotation=90)

sns.lineplot(data = C5ICB, x = C5ICB.index, y = 'spa', color = 'black', marker = 'o')

plt.title('La Liga', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('C5ICB', **font)

plt.ylim(120, 170)
plt.tight_layout()

plt.savefig("spa_C5ICB.png")


fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

#plt.xticks(rotation=90)

sns.lineplot(data = C5ICB, x = C5ICB.index, y = 'ita', color = 'black', marker = 'o')

plt.title('Serie A', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('C5ICB', **font)

plt.ylim(120, 170)
plt.tight_layout()

plt.savefig("ita_C5ICB.png")



fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

#plt.xticks(rotation=90)

sns.lineplot(data = C5ICB, x = C5ICB.index, y = 'ger', color = 'black', marker = 'o')

plt.title('Bundesliga', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('C5ICB', **font)

plt.ylim(120, 170)
plt.tight_layout()

plt.savefig("ger_C5ICB.png")



fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

#plt.xticks(rotation=90)

sns.lineplot(data = C5ICB, x = C5ICB.index, y = 'fra', color = 'black', marker = 'o')

plt.title('Ligue 1', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('C5ICB', **font)

plt.ylim(120, 170)
plt.tight_layout()

plt.savefig("fra_C5ICB.png")
```
