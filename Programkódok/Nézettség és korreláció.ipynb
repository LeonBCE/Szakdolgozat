# Átlagos éves nézőszámok a top 5 bajnokságban 2000 és 2023 között

Programcsomagok importálása

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Lokális adatbázis importálása

```python
att_loc = pd.read_csv('A top 5 bajnokság 2000-2023 közötti éves átlagos nézőszámadatai.csv')
```

```python
att_loc = att_loc[1:]
att_loc.reset_index(drop = True, inplace = True)
att_loc.head(5)
```

Nézettségi ábrák elkészítése és exportálása

```python
font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

sns.lineplot(data = att_loc, x = att_loc['year'], y = 'eng', color = 'black', marker = 'o')

plt.title('Premier League', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Átlagos nézőszám', **font)


plt.tight_layout()

plt.savefig("pics/eng_att.png")



font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

sns.lineplot(data = att_loc, x = att_loc['year'], y = 'ita', color = 'black', marker = 'o')

plt.title('Serie A', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Átlagos nézőszám', **font)

plt.tight_layout()

plt.savefig("pics/ita_att.png")




font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

sns.lineplot(data = att_loc, x = att_loc['year'], y = 'spa', color = 'black', marker = 'o')

plt.title('La Liga', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Átlagos nézőszám', **font)

plt.tight_layout()

plt.savefig("pics/spa_att.png")




font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

sns.lineplot(data = att_loc, x = att_loc['year'], y = 'ger', color = 'black', marker = 'o')

plt.title('Bundesliga', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Átlagos nézőszám', **font)

plt.tight_layout()

plt.savefig("pics/ger_att.png")




font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

sns.lineplot(data = att_loc, x = att_loc['year'], y = 'fra', color = 'black', marker = 'o')

plt.title('Ligue 1', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Átlagos nézőszám', **font)

plt.tight_layout()

plt.savefig("pics/fra_att.png")
```
