# Dobozábra a top 5 bajnokság szezonvégi gólkülönbségeiről, 2000 és 2023 között

Programcsomagok importálása

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Adatok beolvasása csv fájlból

```python
df = pd.read_csv('A top 5 bajnokság 2000-2023 közötti adatai.csv')
df2 = df
```

Adatok szétbontása bajnokságonként

```python
eng_df = df2.loc[df2['League'] == 'Premier-League']
spa_df = df2.loc[df2['League'] == 'La-Liga']
ita_df = df2.loc[df2['League'] == 'Serie-A']
ger_df = df2.loc[df2['League'] == 'Bundesliga']
fra_df = df2.loc[df2['League'] == 'Ligue-1']
```

Dobozábrák vizualizálása, majd exportálása png fájlként

```python
font = {'fontname' : 'Times New Roman'}

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

plt.xticks(rotation=90)

sns.boxplot(data = eng_df, x = 'Season', y = 'GD', color = 'grey', linecolor = 'black')

plt.title('Premier League', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Gólkülönbség', **font)

plt.ylim(-100, 100)
plt.tight_layout()

plt.savefig("pics/eng_boxplot_gd.png")


fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

plt.xticks(rotation=90)

sns.boxplot(data = ita_df, x = 'Season', y = 'GD', color = 'grey', linecolor = 'black')


plt.title('Serie A', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Gólkülönbség', **font)

plt.ylim(-100, 100)
plt.tight_layout()


plt.savefig("pics/ita_boxplot_gd.png")

fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

plt.xticks(rotation=90)

sns.boxplot(data = spa_df, x = 'Season', y = 'GD', color = 'grey', linecolor = 'black')


plt.title('La Liga', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Gólkülönbség', **font)

plt.ylim(-100, 100)
plt.tight_layout()


plt.savefig("pics/spa_boxplot_gd.png")


fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

plt.xticks(rotation=90)

sns.boxplot(data = ger_df, x = 'Season', y = 'GD', color = 'grey', linecolor = 'black')


plt.title('Bundesliga', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Gólkülönbség', **font)

plt.ylim(-100, 100)
plt.tight_layout()

plt.savefig("pics/ger_boxplot_gd.png")


fig = plt.figure(figsize = (5,5), linewidth = 1)

sns.set_style("ticks", {'axes.grid' : False})

plt.xticks(rotation=90)

sns.boxplot(data = fra_df, x = 'Season', y = 'GD', color = 'grey', linecolor = 'black')


plt.title('Ligue 1', **font, fontweight = 'bold')
plt.xlabel('Szezon', **font)
plt.ylabel('Gólkülönbség', **font)

plt.ylim(-100, 100)
plt.tight_layout()

plt.savefig("pics/fra_boxplot_gd.png")
```
