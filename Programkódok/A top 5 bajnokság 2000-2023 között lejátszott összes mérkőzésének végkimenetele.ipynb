# A top 5 bajnokság 2000 és 2023 között lejátszott összes mérkőzéseredménye

Programcsomagok importálása

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
```

Webscraping az https://fbref.com oldaláról

```python
start_year = 2000
end_year = 2023
fields = ['Wk', 'Date', 'Home', 'Score', 'Away']
leagues = [('Premier-League', 9), ('La-Liga', 12), ('Serie-A', 11), ('Bundesliga', 20), ('Ligue-1', 13)]

lst = []

for league_name, league_id in leagues:
    for i in range(start_year, end_year):
        BASE ='https://fbref.com'
        URL = f'{BASE}/en/comps/{league_id}/{i}-{i+1}/schedule/{i}-{i+1}-{league_name}-Scores-and-Fixtures'
        print(URL)
        df = pd.read_html(URL)[0]
        df = df[fields]
        df = df.reset_index(drop=True)
        df.insert(0, "League", league_name)
        df.insert(0, "Season", i+1)
        lst.append(df)
        
        time.sleep(60/20)
```

Adatok tisztítása, felesleges oszlopok ejtése, hazai és idegenbeli gólszám, gólkülönbség, gólok számának kiszámítása és számformátummá alakítása, majd exportálás csv fájlba

```python
big_df = pd.concat(lst, axis=0, ignore_index = True)
big_df = big_df[big_df['Score'].notna()]
big_df = big_df[big_df['Wk'].notna()]

big_df['Home_score'] = big_df['Score'].str.split('–').str[0].astype('int')
big_df['Away_score'] = big_df['Score'].str.split('–').str[1].astype('int')
big_df['GD'] = abs(big_df['Home_score'] - big_df['Away_score'])
big_df['Sum_G'] = big_df['Home_score'] + big_df['Away_score']

big_df.to_csv(f'A top 5 bajnokság {start_year}-{end_year} között lejátszott összes mérkőzésének végkimenetele.csv', encoding = 'utf-8', index = False)

big_df.head(5)
```

```python
eng_df = big_df.loc[big_df['League'] == 'Premier-League'].reset_index(drop = True)
spa_df = big_df.loc[big_df['League'] == 'La-Liga'].reset_index(drop = True)
ita_df = big_df.loc[big_df['League'] == 'Serie-A'].reset_index(drop = True)
ger_df = big_df.loc[big_df['League'] == 'Bundesliga'].reset_index(drop = True)
fra_df = big_df.loc[big_df['League'] == 'Ligue-1'].reset_index(drop = True)
```
